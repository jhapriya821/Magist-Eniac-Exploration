```python
import pandas as pd

# Load the data into a table (DataFrame)
df = pd.read_csv('marketing_campaign_dataset.csv')

# Display the first 5 rows
print("--- Data Preview ---")
df.head()
```

```python
# 1. See what columns we have
print("Columns in our dataset:", df.columns.tolist())

# 2. Check the average ROI or Conversion by Campaign Type
# (Note: I'm assuming column names like 'Campaign_Type' - the code below will show yours)
summary = df.groupby('Campaign_Type').size()
print("\nNumber of Campaigns by Type:")
print(summary)

# 3. Create a simple chart to show your boss
import matplotlib.pyplot as plt
summary.plot(kind='bar', color='skyblue')
plt.title('Campaign Distribution')
plt.show()
```
